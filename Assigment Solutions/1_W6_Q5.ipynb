```python
import numpy as np
import pandas as pd
import scipy
import math
from scipy import stats
import statsmodels.api as spm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

## 5) Without Blocking

```python
a = [75,76,78]
b = [74,74,74]
c = [70,71,75]
d = [73,72,77]
e = [76,73,76]
f = [73,73,73]
stats.f_oneway(a, b, c, d, e, f)
```

```python
data = {'S1':[75,74,70,73,76,73], 'S2':[76,74,71,72,73,73], 'S3':[78,74,75,77,76,73]}
df = pd.DataFrame(data)
df
```

```python
dfnew = pd.melt(df.reset_index(), id_vars = ['index'], value_vars = ['S1', 'S2', 'S3'])
dfnew.columns = ['index', 'Treatments', 'value']
dfnew
```

```python
model = ols('value ~ C(Treatments)', data = dfnew).fit()
anova_table = spm.stats.anova_lm(model, typ=1)
anova_table
```

## 6) With Blocking

```python
dfnew = pd.melt(df.reset_index(), id_vars = ['index'], value_vars = ['S1', 'S2', 'S3'])
dfnew.columns = ['blocks', 'Treatments', 'value']
```

```python
model = ols('value ~ C(blocks) + C(Treatments)', data = dfnew).fit()
anova_table = spm.stats.anova_lm(model, typ = 1) 
anova_table
```
