```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df2 = pd.read_excel('Twoway.xlsx')
df2
```

```python
formula = 'Value ~C(college) + C(prep_pro) + C(college) : C(prep_pro)'
model = ols(formula, df2).fit()
aov_table = sm.stats.anova_lm(model, typ = 2) #tpy = 2, 2 way anova
print(aov_table)
```
