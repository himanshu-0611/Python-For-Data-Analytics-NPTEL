```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl                         #Plot figures
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from scipy import stats
from statsmodels.stats.anova import anova_lm     #O/P of anova table of reg model
from statsmodels.formula.api import ols          #for regression analysis
```

```python
data = {'A':[58,64,55,66,67], 'B':[58,69,71,64,68], 'C':[48,57,59,47,49]}
data = pd.DataFrame(data)
data
```

```python
data_r = pd.melt(data.reset_index(), id_vars = ['index'], value_vars = ['A', 'B', 'C'])
data_r.columns = ['index', 'treatments', 'value']
data_r
```

```python
model = ols('value ~ C(treatments)', data = data_r).fit() #value = dep var
anova_table = sm.stats.anova_lm(model, typ = 1) #lm - linear model, typ = 1, type 1 ANOVA
anova_table
```

```python
just_dummies = pd.get_dummies(data_r['treatments'])
just_dummies
```

```python
step_1 = pd.concat([data_r, just_dummies], axis = 1)
step_1.drop(['treatments', 'C'], inplace = True, axis = 1)
step_1
```

```python
result = smf.ols(step_1['value'], sm.add_constant(step_1[['A', 'B']])).fit()
result.summary()
```

## <font color='red'> Intercept (beta0) = 52, A (beta1) = 10, B (beta2) = 14<br>
<b> F = 0.0038 i.e. Reject Null Ho : beta0 != beta1 != beta2

```python
model = ols('value ~ A+B', data=step_1).fit()
anova_table = sm.stats.anova_lm(model, typ = 1)
anova_table
```
